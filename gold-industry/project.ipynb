## Описание данных
### Технологический процесс
- *Rougher feed* — исходное сырье
- *Rougher additions* (или *reagent additions*) — флотационные реагенты: *Xanthate, Sulphate, Depressant*
    - *Xanthate* — ксантогенат (промотер, или активатор флотации);
    - *Sulphate* — сульфат (на данном производстве сульфид натрия);
    - *Depressant* — депрессант (силикат натрия).
- *Rougher process* (англ. «грубый процесс») — флотация
- *Rougher tails* — отвальные хвосты
- *Float banks* — флотационная установка
- *Cleaner process* — очистка
- *Rougher Au* — черновой концентрат золота
- *Final Au* — финальный концентрат золота
### Параметры этапов
- *air amount* — объём воздуха
- *fluid levels* — уровень жидкости
- *feed size* — размер гранул сырья
- *feed rate* — скорость подачи
### Наименование признаков
Наименование признаков должно быть такое:
`[этап].[тип_параметра].[название_параметра]`

Пример: `rougher.input.feed_ag`

Возможные значения для блока `[этап]`:
- *rougher* — флотация
- *primary_cleaner* — первичная очистка
- *secondary_cleaner* — вторичная очистка
- *final* — финальные характеристики

Возможные значения для блока `[тип_параметра]`:
- *input* — параметры сырья
- *output* — параметры продукта
- *state* — параметры, характеризующие текущее состояние этапа
- *calculation* — расчётные характеристики

### Проверка расчета эффективности обогащения

```python
def recovery(row):
    # доля золота в сырье/концентрате до флотации/очистки
    f = row['rougher.input.feed_au']
    # доля золота в концентрате после флотации/очистки
    c = row['rougher.output.concentrate_au']
    # доля золота в отвальных хвостах после флотации/очистки
    t = row['rougher.output.tail_au']
    
    return c * (f - t) * 100 / (f * (c - t))
```

```python
# Флотация
data_train['rougher.output.recovery_c'] = data_train.apply(recovery, axis=1)
```

```python
mean_absolute_error(data_train['rougher.output.recovery'], data_train['rougher.output.recovery_c'])
```

**Вывод**: эффективность обогащения рассчитана правильно, т.к. метрика MAE пренебрежима мала.

Проведем двусторонний статистический тест на равенство средних двух независимых выборок.
- Нулевая гипотеза $H_0$: средние не различаются.

## Модель
Как показано ниже, в признаках присутствует мультиколлинеарность. Об этом говорят стремящиеся к нулю определители матриц корреляции признаков, а также высокая корреляция между некоторыми признаками. На каждом этапе (черновой и финальный) будут предприняты шаги по уменьшению мультиколлинеарности:
- удаляются признаки, которые имеют слабую корреляцию с целевым признаком;
- из пар высоко коррелированных признаков удаляется один, который имеет минимиальную корреляцию с целевым признаком;
- используется модель с регуляризацией.

```python
def smape(y_true, y_pred):
    '''
    Метрика sMAPE
    '''
    sm = 2 * sum(
        abs(y_true - y_pred) / (abs(y_true) + abs(y_pred))
    )
    return sm * 100 / len(y_true)
```

```python
def overall_smape(rougher_smape, final_smape):
    '''
    Итоговое sMAPE
    '''
    return 0.25 * rougher_smape + 0.75 * final_smape
```

```python
def grid_cv(X, y, model, params, scoring, cv=5, n_jobs=1, random=False, n_iter=25, verbose=3):
    '''
    Подбор гиперпатаметров по сетке с кросс-валидацией
    Если параметр random = True, то будет использоваться случайный поиск по сетке
    '''
    if random:
        grid = RandomizedSearchCV(
            model,
            params,
            n_iter=n_iter,
            scoring=smape_scorer,
            cv=cv,
            verbose=verbose,
            random_state=RANDOM_STATE,
            n_jobs=n_jobs)
    else:
        grid = GridSearchCV(
            estimator=model,
            param_grid=params,
            scoring=scoring,
            cv=cv,
            n_jobs=n_jobs)
    
    # Fit the gridsearch
    grid.fit(X, y)
    
    # Скорер создан с параметром greater_is_better=False
    # Кросс-валидация с таким параметром умножает на -1 значения метрики (sMAPE)
    # Знак минус перед best_score_ делает метрику вновь положительной
    print(f'Лучшая метрика sMAPE: {-grid.best_score_:.2f}')
    print(f'Лучший набор гиперпараметров: {grid.best_params_}')

    return grid.best_estimator_
```

```python
def describe_cols(data, cols, display_cols=False):
    '''
    Вспомогательная функция для описания столбцов.
    Отображает количество столбцов, по необходимости перечисляет их.
    Также для столбцов считается определитель матрицы корреляции.
    '''
    print(f'Количество столбцов: {len(cols)}')
    if display_cols:
        print(cols)
    print(f'Определитель матрицы корреляции: {np.linalg.det(data[cols].corr())}\n')
```

```python
def remove_corr(data, cols, target_col, upper_threshold=0.9):
    '''
    Рекурсивная функция для удаления столбцов с высоким значением корреляции.
    Алгоритм:
        - находится корреляция признаков (с целевым признаком)
        - выбирается первая пара признаков с корреляцией не меньшей upper_threshold
        - из пары удаляется тот признак, который наименее коррелирован с целевым признаком
        - функция вызывает саму себя для поиска следующей пары
    '''
    inner_corr = data[cols + [target_col]].corr().abs()
    for col in cols:
        s = inner_corr[col]
        s = s[(s >= upper_threshold) & (s.index != col)]
        if len(s):
            idx = s.index[0]
            if inner_corr.loc[col, target_col] < inner_corr.loc[idx, target_col]:
                cols.remove(col)
            else:
                cols.remove(idx)
            remove_corr(data, cols, target_col)
            break
```

```python
def prepare_features(data, stage_cols, stage_target_col, lower_threshold=0.1):
    '''
    Сборная функция объединяющая все этапы подготовки признаков:
    - удаляет пустые значения
    - удаляет столбцы, имеющие слабую корреляцию с целевым признаком
    - удаляет столбцы из сильно коррелированных пар нецелевых признаков
    После каждого этапа отображается определитель матрицы корреляции, для того,
    чтобы продемонстрировать прогресс уменьшения мультиколлинеарности.
    
    Возвращается измененый набор столбцов,
    а также матрица признаков и вектор целевого признака.
    '''
    # Удалим пустые значения перед кросс-валидацией
    data = data.dropna(subset=stage_cols)
    
    # Посмотрим на количество столбцов и определитель корреляционной матрицы
    print('Исходные признаки')
    describe_cols(data, stage_cols)
    
    # Удалим столбцы, которые слабо коррелируют с целевым признаком (меньше 0.1)
    corr = data[stage_cols + [stage_target_col]].corr()
    stage_cols = corr[corr[stage_target_col].abs() >= lower_threshold].index.to_list()
    stage_cols.remove(stage_target_col)
    print('После удаления столбцов с низкой корреляцией с целевым признаком')
    describe_cols(data, stage_cols)
    
    # Удалим сильную корреляцию
    remove_corr(data, stage_cols, stage_target_col)
    print('После удаления столбцов с сильной взаимной корреляцией')
    describe_cols(data, stage_cols, display_cols=True)
    
    X = data[stage_cols]
    y = data[stage_target_col]
    
    return X, y, stage_cols
```

```python
def heatmap(data, stage_cols, stage_target_col, title):
    fig, ax = plt.subplots(figsize=(10, 7))
    _ = sns.heatmap(
        data[stage_cols + [stage_target_col]].corr(),
        vmin=-1,
        vmax=1,
        ax=ax,
        cmap='coolwarm')
    plt.title(title)
```

### Тестирование

```python
def test(cols, target_col, scaler, model):
    '''
    Тестирование
    '''
    # Удаляем пустые значения
    data = data_test.dropna(subset=cols)
    # Получаем матрицу признаков
    X_test = data[cols]
    # Масштабируем признаки
    X_test = scaler.transform(X_test)
    # Получаем вектор целевого признака
    y_test = data[target_col]
    # Предсказываем
    y_pred = model.predict(X_test)
    # Возвращаем метрику
    return smape(y_test, y_pred)
```
